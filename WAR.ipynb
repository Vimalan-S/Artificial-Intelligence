```python
# To stop executing once threshold is reached
import sys 

# To generate Random numbers as Data
import random
random.seed(0)
```

```python
class distinguish_weapons:
    def __init__(self):

        # Setting up the MAXIMUM max_requirement of weapons to be received from other Countries
        self.max_requirement = random.randint(0,100)  

        self.property = {}
        for weapons in range(self.max_requirement):

            # Randomizing the possible Weapons received from other Countries
            self.property[weapons] = random.choice(["Pistol","AK47","Glock17","Rifle","MachineGun","Uzi","Grenade","Sniper","Deagle","ShotGun","M16","FlareGun"])

    def get_max_requirement(self):
        return self.max_requirement

    def get_property(self,weapons):
        return self.property[weapons]
```

```python
# Initializing the Armories

Light_weight_Armory = []
medium_weight_Armory = []
heavy_weight_Armory = []
```

```python
class weapon_distinguishing_Agent:
    def __init__(self):
        self.weapons_count = 0

    def start_to_distinguish_weapons(self,env):
        if self.weapons_count < env.get_max_requirement():

            # If we received a pack of FlareGun/ Glock17/ Deagle/ Uzi , those boxes will be moved to Light-weight Armory.
            if env.get_property(self.weapons_count) =="FlareGun" or env.get_property(self.weapons_count) =="Glock17" or env.get_property(self.weapons_count) =="Deagle" or env.get_property(self.weapons_count) =="Uzi":
                print("Received a pack of ", env.get_property(self.weapons_count))
                print("Moving the item to Light-weight Armory...\n")
                Light_weight_Armory.append(env.get_property(self.weapons_count)) 

                self.weapons_count += 1          

            # If we received a pack of Pistol/ MachineGun/ ShotGun/ AK47 , those boxes will be moved to Medium-weight Armory.
            elif env.get_property(self.weapons_count) =="Pistol" or env.get_property(self.weapons_count) =="MachineGun" or env.get_property(self.weapons_count) =="ShotGun" or env.get_property(self.weapons_count) =="AK47":
                print("Received a pack of ", env.get_property(self.weapons_count))
                print("Moving the item to Medium-weight Armory...\n")
                medium_weight_Armory.append(env.get_property(self.weapons_count))

                self.weapons_count += 1

            # If we received a pack of Rifle/ Grenade/ Sniper/ M16 , those boxes will be moved to Heavy-weight Armory.
            elif env.get_property(self.weapons_count) =="Rifle" or env.get_property(self.weapons_count) =="Grenade" or env.get_property(self.weapons_count) =="Sniper" or env.get_property(self.weapons_count) =="M16":
               print("Received a pack of ", env.get_property(self.weapons_count))
               print("Moving the item to Heavy-weight Armory...\n")
               heavy_weight_Armory.append(env.get_property(self.weapons_count))

            self.weapons_count += 1
            return True 

        else:
            self.weapons_received()
            print("Reached the Maximum number of Weapons required to fight the invading Russians!! \n\nNow Distribution of these Weapons via the Armories is the last task for the Ukranians to defeat the Russians!" )
            return False 

    def weapons_received(self):
        print("\nTotal number of Weapons received from other Countries: ", self.weapons_count)
```

```python
my_agent = weapon_distinguishing_Agent()
my_env = distinguish_weapons()

print("The Maximum number of Weapons needed by the Ukranians = ", my_env.get_max_requirement())

status = True
while(status):
    status = my_agent.start_to_distinguish_weapons(my_env)
```
